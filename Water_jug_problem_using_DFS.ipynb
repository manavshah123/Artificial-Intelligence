```
import queue
import time
import random

x = 0  # jug 1
y = 0  # jug 2
MaxJug1 = (int(input("Enter Jug 1 : "))) # max capacity of jug 1
MaxJug2 = (int(input("Enter Jug 2: ")))  # max capacity of jug 2


# preparing a Queue
DFS_Stack = queue.LifoQueue()


# creating a class
class Node:
    # initializing a constructor for creating a node
    def __init__(self, parent):
        self.x = 0
        self.y = 0
        self.parent = parent

    # to check a goalnode
    def isGoalNode(self, GNode):
        if self.x == GNode:
            return True
        return False

    # print node
    def printNode(self):
        print("(", self.x, ",", self.y, ")")


# generating a successor node
def generateAllSuccessor(cnode, visitednodelist):
    list1 = []
    list_rule = []
    while len(list_rule) < 8:
        Rule = random.randint(1, 8)
        if not Rule in list_rule:
            list_rule.append(Rule)
            next_node = operation(cnode, Rule)
            if next_node is not None and not isnodeinlist(next_node, visitednodelist):
                list1.append(next_node)
    list1.reverse()  # next node
    return list1


def isnodeinlist(node, list1):
    for n in list1:
        if node.x == n.x and node.y == n.y:
            return True
    return False


# checking a rule
def operation(cnode, Rule):
    x = cnode.x
    y = cnode.y

    if Rule == 1:
        if x < MaxJug1:
            x = MaxJug1
        else:
            return None
    elif Rule == 2:
        if y < MaxJug2:
            y = MaxJug2
        else:
            return None
    elif Rule == 3:
        if x > 0:
            x = 0
        else:
            return None
    elif Rule == 4:
        if y > 0:
            y = 0
        else:
            return None
    elif Rule == 5:
        if x + y >= MaxJug1:
            y = y - (MaxJug1 - x)
            x = MaxJug1
        else:
            return None
    elif Rule == 6:
        if x + y >= MaxJug2:
            x = x - (MaxJug2 - y)
            y = MaxJug2
        else:
            return None
    elif Rule == 7:
        if x + y < MaxJug1:
            x = x + y
            y = 0
        else:
            return None
    elif Rule == 8:
        if x + y < MaxJug2:
            y = x + y
            x = 0
        else:
            return None
    else:
        print("invalid input!")

    nextnode = Node(cnode)
    nextnode.x = x
    nextnode.y = y
    nextnode.parent = cnode
    return nextnode


# pushing a node in a list
def PushList(list1):
    for m in list1:
        DFS_Stack.put(m)


# pop a node from a stack
def PopNode():
    if (DFS_Stack.empty()):
        return None
    else:
        return DFS_Stack.get()


# initializing and modifying a dfs algo.
def dfs(initial_node, GNode):
    visitednodelist = []
    DFS_Stack.put(initial_node)
    while not DFS_Stack.empty():
        visited_node = PopNode()
        visitednodelist.append(visited_node)
        if visited_node.isGoalNode(GNode):
            return visited_node
        PushList(generateAllSuccessor(visited_node, visitednodelist))
    return None


# printing a path to the goal node
def PrintPath(CNode):
    temp = CNode
    list2 = []
    while temp is not None:
        list2.append(temp)
        temp = temp.parent
    list2.reverse()
    for i in list2:
        i.printNode()
    print("path:", len(list2))


# calling a class for creating a node
initial_node = Node(None)
currentNode = initial_node
GNode = int(input("enter goal state of jug1:"))
if GNode <= MaxJug1:
    GNode = GNode
else:
    print("capacity must be less than maximum capacity of jug 1!")
    exit()
print("Goal state = ({},y)".format(GNode))
var1 = int(input("enter 1 for manually and enter 2 for dfs:"))

# if 1 for manually writing a all the rules
if var1 == 1:
    while not currentNode.isGoalNode(GNode):
        rule = int(input("enter rule no.:"))
        NextNode = operation(currentNode, rule)
        if NextNode is not None:
            currentNode = NextNode
            currentNode.printNode()

# if 2 for getting a goal state with time!
elif var1 == 2:
    start_time = time.time()
    solution = dfs(initial_node, GNode)
    end_time = time.time()
    if solution is None:
        print("solution not found")
    else:
        PrintPath(solution)
    print("execution time", round((end_time - start_time) * 1000, 2), " millis")

# invalid state if other than 1 or 2!
else:
    print("invalid input")
```
